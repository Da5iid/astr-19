```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from matplotlib import patches
```

```python
f = plt.figure(figsize=(10,14),facecolor='black')

cmap_turbo = cm.get_cmap('turbo')

ax = f.add_subplot(111,xticks=[], yticks=[])
ax.set_axis_off()


#circle = patches.Circle((3.5,10),0.5,edgecolor='None',facecolor='red' ,alpha=1)
#ax.add_patch(circle)

ellipse = patches.Ellipse((3.2,10.5),2,3.45,edgecolor='None',facecolor='red' ,alpha=1)
ax.add_patch(ellipse)

ax.set_xlim([0, 2*np.pi])
ax.set_ylim([0,15.])

x = np.linspace(0,2*np.pi,1000)

nmr = 13

npeak = 13

for i in range(nmr):
    mr_color = cmap_turbo(float(i)/float(npeak))
    
    #rand set of frequencies
    freq = 3.0*np.random.rand(npeak) + 1.0
    
    #rand set of plase shifts
    phi = 2.0*np.pi*np.random.rand(npeak)
    
    #randon amplitude
    amp = 1.0*np.random.rand(npeak) + 0.025
    
    yoff = 0.3*float(i)
    
    y = np.zeros_like(x)
    
    #for every peak
    ##add this peak a sin curve with a 
    #randon amplitude, freq, and phase
    
    for j in range(npeak):
        y += amp[j]*np.sin(freq[j]*x + phi[j])

    y = np.fabs(y) + yoff

    plt.fill_between(x,y,yoff,zorder=nmr-i,color=mr_color,cmap='turbo',alpha=1)



plt.savefig('mountain-range.png',bbox_inches='tight',pad_inches=0,dpi=600,)
    
    
```
