```python
import matplotlib.pyplot as plt
import numpy as np

#needed to install SciPy for the Gamma part.
import scipy.special as sps
```

```python
#defualt normal distribution

n = 1000

x = np.random.randn(n)

width = 0.1
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,color="#f1466f",edgecolor="#ffffff")
plt.xlabel("standard deviations")
plt.ylabel("numbers per bin")
plt.show

plt.savefig('pt10-dist-unif.png',bbox_inches='tight',pad_inches=0,dpi=600,)
```

```python
#numpy.random.Generator.gamma, Code and documentation borrowed from:
    #https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.gamma.html

#From what I can gather, "Gamma Distribution" is part of statistic-probability type fields, and
#according to a Harvard Professor, one of the most famous functions in that field.

shape, scale = 2., 2.
s = np.random.default_rng().gamma(shape, scale, 1000)

count, bins, ignored = plt.hist(s, 100, density=True)
y = bins**(shape-1)*(np.exp(-bins/scale) /
                     (sps.gamma(shape)*scale**shape))

plt.xlabel("sample bins (blue), gamma function (red)")
plt.ylabel("probablility density")
plt.plot(bins, y, linewidth=2, color='r')
plt.show()
```
