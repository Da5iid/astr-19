```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sin(num):
    sinData = []
    x = np.linspace(0.0,2.0*np.pi,num)
    
    for i in range(num):
        sinData.append(np.sin(x[i]))
    
    return sinData
```

```python
def cos(num):
    cosData = []
    x = np.linspace(0.0,2.0*np.pi,num)
    
    for i in range(num):
        cosData.append(np.cos(x[i]))
    
    return cosData
```

```python
num = 100
x = np.linspace(0,1,num)

sinx = sin(num)
cosx = cos(num)

f, charts = plt.subplots(1, 2)

charts[0].plot(x, sinx)
charts[0].set_title(r'$\sin(x)$')

charts[1].plot(x, cosx)
charts[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace = 0.5)
```
