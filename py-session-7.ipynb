```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exponent(arr):
    return np.exp(arr)
```

```python
x = np.linspace(0.0,1.0,100)
y = exponent(x)

f = plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.xlabel('Time (miliseconds)')
plt.ylabel("Awesomeness")

plt.savefig('pysession7.png',bbox_inches="tight",dpi=1200,)
plt.show()
```
